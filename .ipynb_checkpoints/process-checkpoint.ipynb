```python
def get_data(n):
    data_dir = 'data/%02d_input' % n
    img = plt.imread(join(data_dir, 'img.png')) 
    g_row, g_col = map(int, open(join(data_dir, 'g_coord.csv')).read().rstrip('\n').split(','))
    h = img.shape[0] // 3
    b_row, b_col = g_row + h, g_col
    r_row, r_col = g_row - h, g_col
    return img, (g_row, g_col), (b_row, b_col), (r_row, r_col)
```

```python
def trim(img):
    dh, dw = map(lambda x: int(x * 0.3), img.shape[:2])
    return img[dh:-dh, dw:-dw]

def split(img):
    h = img.shape[0] // 3
    return list(map(trim, (img[i * h:(i + 1) * h] for i in range(3))))
```

```python
def mse(a, b):
    return ((a - b) ** 2).mean()

def cross(a, b):
    return -(a * b).sum() / ((a * a).sum() * (b * b).sum()) ** 0.5
```

```python
def shift2(a, b, x, y, n, scorer=mse):
    best_score, best_dx, best_dy = 100, 0, 0
    for dx in range(-n + x, n + x + 1):
        for dy in range(-n + y, n + y + 1):
            p = -abs(dx) if dx != 0 else a.shape[0]
            q = -abs(dy) if dy != 0 else a.shape[1]
            score = scorer(np.roll(a, (dx, dy), (0, 1))[:p, :q],
                           b[:p, :q])
            if score < best_score:
                best_score, best_dx, best_dy = score, dx, dy
    return best_dx - x, best_dy - y
```

```python
def shift(a, b, n): #shift a relative to b
    best_score = mse(a, b)
    best_dx = 0
    best_dy = 0
    for dx in range(1, n + 1):
        for dy in range(1, n + 1):
            score = mse(a[dx:, dy:], b[:-dx, :-dy])
            print(score)
            if score < best_score:
                best_score, best_dx, best_dy = score, dx, dy
                
    for dy in range(1, n + 1):
        score = mse(a[:, dy:], b[:, :-dy])
        if score < best_score:
            best_score, best_dx, best_dy = score, 0, dy
            
    for dx in range(1, n + 1):
        score = mse(a[dx:], b[:-dx])
        if score < best_score:
            best_score, best_dx, best_dy = score, dx, 0
            
    return best_dx, best_dy

def dcoord(b, a, n, shift=shift):
    score1, dx1, dy1 = shift(a, b, n)
    score2, dx2, dy2 = shift(b, a, n)
    if score1 < score2:
        dx, dy = dx1, dy1
    else:
        dx, dy = -dx2, -dy2
    return dx, dy
```

```python
def align(img, gc):
    h = img.shape[0] // 3
    
    gc = np.array(gc)
    rc = gc + (h, 0)
    bc = gc - (h, 0)
    
    bd = np.zeros(2, dtype=np.int32)
    rd = np.zeros(2, dtype=np.int32)
    
    if h > 1000:
        k, cnt = 3, 2
        
        b, g ,r = map(lambda x: list(pyramid_gaussian(x, cnt, k)), split(img))
        
        for i in range(cnt + 1):
            n = k ** (cnt - i) + 2
            rd *= k
            bd *= k
            
            bd += shift2(g[cnt - i], b[cnt - i], bd[0], bd[1], n)        
            rd += shift2(g[cnt - i], r[cnt - i], rd[0], rd[1], n)
        
        bc += bd
        rc += rd
        
        ret = np.dstack((np.roll(b[0], (-bd), (0, 1)), 
                         g[0],
                         np.roll(r[0], (-rd), (0, 1))))
                        
    else:
        n = 15
        
        b, g ,r = split(img)
        
        bd += shift2(g, b, n)
        rd += shift2(g, r, n)
        
        bc += bd
        rc += rd
        
        ret = np.dstack((np.roll(b, (bd[0], bd[1]), (0, 1)), 
                         g,
                         np.roll(r, (rd[0], rd[1]), (0, 1))))
                        
    return ret, (bc[0], bc[1]), (rc[0], rc[1])
```
